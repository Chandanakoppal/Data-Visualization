```python
import numpy as np
from pandas import DataFrame
import pandas as pd
```

```python
df = pd.read_csv('milk_production.csv')
df = df.applymap(lambda x: x.strip() if isinstance(x,str) else x)
df = df.dropna()
```

```python
df.head(10)
```

```python
# Which state has produced max milk in year 2013-14?

df['milk'] = df['Cow Milk-2013-14'] + df['Boffalo Milk-2013-14'] + df['Goat Milk-2013-14']
df[df['milk'] == df['milk'].max()] [['State/ UT Name']]
```

```python
# List top 5 milk producing states in each year from 2010 to 2015

df['10-11'] = df['Cow Milk-2010-11'] + df['Boffalo Milk-2010-11'] + df['Goat Milk-2010-11']
df['11-12'] = df['Cow Milk-2011-12'] + df['Boffalo Milk-2011-12'] + df['Goat Milk-2011-12']
df['13-14'] = df['Cow Milk-2013-14'] + df['Boffalo Milk-2013-14'] + df['Goat Milk-2013-14']
df['14-15'] = df['Cow Milk-2014-15'] + df['Boffalo Milk-2014-15'] + df['Goat Milk-2014-15']

milk = df.sort_values(['10-11','11-12','13-14','14-15'],ascending = [False,False,False,False])

milk.head()['State/ UT Name']


```

```python
# Calculate average milk production by cow, buffalo and goat in each year for selected state. 

state=input('Enter the state')

df['mean_cow'] = sum(df.iloc[:,3:7].mean(axis=1))
df['mean_buffalo'] = sum(df.iloc[:,8:12].mean(axis=1))
df['mean_goat'] = sum(df.iloc[:,13:18].mean(axis=1))            

s = df[df['State/ UT Name'] == state][['State/ UT Name']]
```

```python
# List 5 states (if present) whose total milk production is increased for last three years.

df['10-11'] = df['Cow Milk-2010-11'] + df['Boffalo Milk-2010-11'] + df['Goat Milk-2010-11']
df['11-12'] = df['Cow Milk-2011-12'] + df['Boffalo Milk-2011-12'] + df['Goat Milk-2011-12']
df['13-14'] = df['Cow Milk-2013-14'] + df['Boffalo Milk-2013-14'] + df['Goat Milk-2013-14']
df['14-15'] = df['Cow Milk-2014-15'] + df['Boffalo Milk-2014-15'] + df['Goat Milk-2014-15']
df['15-16'] = df['Cow Milk-2014-15'] + df['Boffalo Milk-2014-15'] + df['Goat Milk-2014-15']

df[(df['15-16']>df['14-15']) & (df['14-15']>df['13-14']) & (df['13-14']>df['11-12'])][['State/ UT Name','11-12','13-14','14-15','15-16']].head()
```

```python
#Draw line graph showing total milk production for last 5 years of selected state

import matplotlib.pyplot as plt
import pandas as pd

plt.plot(df['10-11'], df.iloc[2:3], linestyle="dotted", marker='^', color="red", label = '2010')

plt.plot(df['11-12'], df.iloc[2:3], linestyle="solid", marker='o', color="green", label = '2011')

#plt.plot(df['12-13'], df['Andhra Pradesh'], linestyle="dashed", marker='<', color="blue", label = '2012')

plt.xlabel("Year")

plt.ylabel("production")

plt.legend()

plt.show()
```
